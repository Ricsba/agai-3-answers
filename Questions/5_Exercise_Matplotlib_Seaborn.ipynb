**Q-14** Build a corelation of all predictors with 'Churn' and plot a bar chart (Use dataframe.corr() function)

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(10, 8))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt='.2f', linewidths=0.5)
plt.title('Correlation with Churn')
plt.show()
```

**Q-15** Plot a pie chart showing the distribution of the Payment Methods

**Q-16** Plot a line chart showing the distribution of tenure

```python
df['tenure'].value_counts().sort_index().plot(kind='line', figsize=(10, 6))
plt.title('Distribution of Tenure')
plt.xlabel('Tenure')
plt.ylabel('Count')
plt.show()
```

**Q-17** Plot an area chart showing the distribution of tenure

```python
df.groupby('tenure').size().plot(kind='area', figsize=(10, 6))
plt.title('Distribution of Tenure')
plt.xlabel('Tenure')
plt.ylabel('Count')
plt.show()
```
